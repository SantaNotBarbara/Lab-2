```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt,exp

```

```python
img = cv2.imread('zebra.jpg')

#SHOWING AN ORIGINAL IMAGE (GRAY)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, 'gray')
```

```python
#REMOVING NOISE
kernel = np.ones((3,3),np.float32)/9
smooth_img = cv2.filter2D(gray,-1,kernel)
plt.imshow(smooth_img, 'gray')
```

```python
#GAUSSIAN NOISE
img = cv2.imread('zebra.jpg')
gaussian_noise = np.random.normal(0,1,img.size)
gaussian_noise = gaussian_noise.reshape(img.shape[0], img.shape[1], img.shape[2]).astype('uint8')
img_gauss = cv2.add(img, gaussian_noise)
plt.imshow(img_gauss,'gray')
```

```python
#Speckle Noise Generation
speckle_noise = img + img * gaussian_noise
gray_s = cv2.cvtColor(speckle_noise, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_s,'gray')
```

```python
#Using Scikit-Image (salt and pepper)

from skimage.util import random_noise
noise_img = random_noise(img, mode='s&p', amount=0.05)
noise_img = np.array(noise_img * 255, dtype=np.uint8)
plt.figure(figsize=(30, 30))
plt.imshow(np.hstack([cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cv2.cvtColor(noise_img, cv2.COLOR_BGR2RGB)]))
```

```python
from scipy.misc import face
from scipy.signal.signaltools import wiener
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng()
img = rng.random((40, 40))  
filtered_img = wiener(img, (5, 5)) 
f, (plot1, plot2) = plt.subplots(1, 2)
plot1.imshow(img)
plot2.imshow(filtered_img)
plt.show()
```
